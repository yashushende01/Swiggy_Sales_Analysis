```python
plt.figure(figsize=(7, 5))

A = sns.countplot(x=df["City"], palette="Set2")
for i in A.containers:
    A.bar_label(i)
    
plt.title("Distribution of City")
plt.xlabel("city")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

A = sns.countplot(x=df["Avg ratings"], palette="Set1")
for i in A.containers:
    A.bar_label(i)
    
plt.title("Distribution of Avg ratings")
plt.xlabel("ratings")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# conclusion

  - Delhi (3.53) has the lowest average rating, which could indicate high competition or inconsistent quality.

- **Pricing Trends:**

  - Mumbai has the most expensive restaurants, followed by Bangalore and Kolkata.
